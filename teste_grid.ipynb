```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import RegularPolygon
import random

# 1. Definição das cores dos tipos de terrenos
tipos = {
    "lago" : "#5199D2",
    "montanha": "#B9BDC2",
    "floresta": "#6FB76F",
    "pantano": "#796085",
    "deserto": "#FED957" 
}

# 2. Criar a lista com a quantidade exata de cada terreno (total = 108)
lista_terrenos = (
    ["lago"] * 22 +
    ["montanha"] * 22 +
    ["floresta"] * 22 +
    ["pantano"] * 21 +
    ["deserto"] * 21
)

# 3. Embaralhar aleatoriamente os terrenos
random.seed(42) # Defina uma semente opcional para manter o mesmo mapa toda vez que rodar, remova se quiser 100% aleatório a cada execução
random.shuffle(lista_terrenos)

# --- Seu código original adaptado ---

altura_hex = 2
r = altura_hex / np.sqrt(3) # raio do hexágono

fig, ax = plt.subplots(figsize=(14, 10)) # tamanho da figura ligeiramente maior para melhor visualização

# totalizando 108 hexágonos
linhas = 9 # quantidade de linhas do grid
colunas = 12 # quantidade de colunas do grid 

contador = 0
for col in range(colunas):
    for row in range(linhas):

        x = col * 1.5 * r
        y = row * np.sqrt(3) * r

        if col % 2:
            y += (np.sqrt(3) / 2) * r

        # Pega o tipo de terreno atual da lista embaralhada e busca a cor correspondente
        terreno_atual = lista_terrenos[contador]
        cor_atual = tipos[terreno_atual]
        contador += 1

        hexagono = RegularPolygon(
            (x, y),
            numVertices=6,
            radius=r,
            orientation=np.radians(30),
            edgecolor="black",
            facecolor=cor_atual,   # Cor aplicada aqui
            linewidth=1.2
        )

        ax.add_patch(hexagono)
        
        # Texto mostrando a coordenada (Linha, Coluna)
        ax.text(x, y, f"({row+1}, {col+1})", ha="center", va="center", fontsize=9, fontweight="medium")

# Ajustes gráficos
ax.set_aspect("equal")
ax.invert_yaxis()
ax.autoscale_view()
plt.axis("off") # Oculta os eixos numéricos para o mapa ficar limpo

# Adicionando uma legenda explicativa no canto superior direito
elementos_legenda = [plt.Line2D([0], [0], marker='h', color='w', label=f"{terreno.capitalize()}",
                                markerfacecolor=cor, markersize=14, markeredgecolor='black')
                     for terreno, cor in tipos.items()]
ax.legend(handles=elementos_legenda, loc='upper right', bbox_to_anchor=(1.12, 1))
plt.show()
```

## Peça do mapa individual

```python
altura_hex = 2
r = altura_hex / np.sqrt(3) # raio do hexágono

fig, ax = plt.subplots(figsize=(12, 8)) # tamanho da figura

# totalizando 18 hexágonos
linhas = 3 # quantidade de linhas do grid
colunas = 6 # quantidade de colunas do grid 

for col in range(colunas):
    for row in range(linhas):

        x = col * 1.5 * r
        y = row * np.sqrt(3) * r

        if col % 2:
            y += (np.sqrt(3) / 2) * r

        hexagono = RegularPolygon(
            (x, y),
            numVertices=6,
            radius=r,
            orientation=np.radians(30),
            edgecolor="black",
            facecolor="lightgray",
        )

        ax.text(x, y, f"({row+1}, {col+1})", ha="center",)
        ax.add_patch(hexagono)

ax.set_aspect("equal")
ax.invert_yaxis() # invertendo o eixo y
ax.autoscale_view()
plt.show()
```
