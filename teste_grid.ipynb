```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import RegularPolygon
```

```python
pecas_mapas = {
    # indica a quantidade de cada tipo de terreno existente em uma peça de mapa; cada peça possui 18 hexágonos no total
    "m1" : {"l": 5, "m": 0, "f": 5, "p": 4, "d": 4},
    "m2" : {"l": 0, "m": 4, "f": 6, "p": 4, "d": 4},
    "m3" : {"l": 5, "m": 5, "f": 4, "p": 4, "d": 0},
    "m4" : {"l": 3, "m": 5, "f": 3, "p": 0, "d": 7},
    "m5" : {"l": 5, "m": 5, "f": 0, "p": 4, "d": 4},
    "m6" : {"l": 4, "m": 3, "f": 4, "p": 5, "d": 2},
}
```

```python
tipos = {
    # tipos de terrenos possíveis no mapa (representados por uma cor em específico)
    "lago" : "#5199D2",
    "montanha": "#B9BDC2",
    "floresta": "#6FB76F",
    "pantano": "#796085",
    "deserto": "#FED957" 
}
```

## Peça do mapa individual

```python
altura_hex = 2
r = altura_hex / np.sqrt(3) # raio do hexágono

fig, ax = plt.subplots(figsize=(12, 8)) # tamanho da figura

# totalizando 18 hexágonos
linhas = 3 # quantidade de linhas do grid
colunas = 6 # quantidade de colunas do grid 

for col in range(colunas):
    for row in range(linhas):

        x = col * 1.5 * r
        y = row * np.sqrt(3) * r

        if col % 2:
            y += (np.sqrt(3) / 2) * r

        hexagono = RegularPolygon(
            (x, y),
            numVertices=6,
            radius=r,
            orientation=np.radians(30),
            edgecolor="black",
            facecolor="lightgray",
        )

        ax.text(x, y, f"({row+1}, {col+1})", ha="center",)
        ax.add_patch(hexagono)

ax.set_aspect("equal")
ax.invert_yaxis() # invertendo o eixo y
ax.autoscale_view()
plt.show()
```

## Mapa com todas as peças (simulação)

```python
altura_hex = 2
r = altura_hex / np.sqrt(3) # raio do hexágono

fig, ax = plt.subplots(figsize=(12, 8)) # tamanho da figura

# totalizando 108 hexágonos
linhas = 9 # quantidade de linhas do grid
colunas = 12 # quantidade de colunas do grid 

for col in range(colunas):
    for row in range(linhas):

        x = col * 1.5 * r
        y = row * np.sqrt(3) * r

        if col % 2:
            y += (np.sqrt(3) / 2) * r

        hexagono = RegularPolygon(
            (x, y),
            numVertices=6,
            radius=r,
            orientation=np.radians(30),
            edgecolor="black",
            facecolor="lightgray"
        )

        ax.add_patch(hexagono)
        ax.text(x, y, f"({row+1}, {col+1})", ha="center",)

ax.set_aspect("equal")
ax.invert_yaxis()
ax.autoscale_view()
plt.show()
```
